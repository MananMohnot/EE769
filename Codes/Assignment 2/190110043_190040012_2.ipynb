```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import sklearn
```

```python
url_train="https://www.ee.iitb.ac.in/~asethi/Dump/MouseTrain.csv" #referencing url of data
df_train = pd.read_csv(url_train) #importing train data
display(df_train) # display train data

url_test = "https://www.ee.iitb.ac.in/~asethi/Dump/MouseTest.csv" #referencing url of test data
df_test = pd.read_csv(url_test) #importing test data
# display(df_test) # display test data
```

```python
def Drop_Col (df_train,df_test,threshold = 0.25): #drops column which have nan values above certain threshold for both test and train
  number_na = [df_train[x].isna().sum() for x in df_train.columns] #finding number of nan per column
  number_na = np.array(number_na) #converting list to array
  index_dropped = np.where(number_na > threshold*np.shape(df_train)[0])[0] #finding index of columns to be dropped
  df_train.drop(df_train.columns[index_dropped], axis = 1, inplace = True) #dropping columns in train
  df_test.drop(df_test.columns[index_dropped], axis = 1, inplace = True) #dropping columns in test
```

```python
Drop_Col(df_train,df_test)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler
```

```python
def imputation(df,strat='mean'): #imputation using given strategy, default mean
  number_na = [df[x].isna().sum() for x in df.columns] #finding number of nan per column
  number_na = np.array(number_na) #converting list to array
  index_dropped = np.where(number_na > 0)[0] #finding index of columns to be imputed
  imputer = SimpleImputer(strategy=strat, missing_values=np.nan) #imputation of strategy in nan
  imputer = imputer.fit(df.iloc[:,index_dropped]) #fit
  df.iloc[:,index_dropped] = imputer.transform(df.iloc[:,index_dropped]) #transform
```

```python
imputation(df_train)
imputation(df_test)
```

```python
def scale(df):
  scaler = RobustScaler()
  df.iloc[:,1:-3] = scaler.fit_transform(df.iloc[:,1:-3])
```

```python
scale(df_train)
scale(df_test)
```

```python
X = df_train.iloc[:,1:-3]
y = df_train.iloc[:,-2:-1]
X_test = df_test.iloc[:,1:-3]
y_test = df_test.iloc[:,-2:-1]
from sklearn import preprocessing
label1 = preprocessing.LabelEncoder()
label2 = preprocessing.LabelEncoder()
label1_test = preprocessing.LabelEncoder()
label2_test = preprocessing.LabelEncoder()
y['Genotype']=label1.fit_transform(df_train['Genotype'])
y['Treatment_Behavior']=label2.fit_transform(df_train['Treatment_Behavior'])
y_test['Genotype']=label1_test.fit_transform(df_test['Genotype'])
y_test['Treatment_Behavior']=label2_test.fit_transform(df_test['Treatment_Behavior'])
```

```python
from imblearn.over_sampling import SMOTE
oversampler1 = SMOTE()
oversampler2 = SMOTE()
oversampler1_test = SMOTE()
oversampler2_test = SMOTE()
X_over1, y_over1 = oversampler1.fit_resample(X, y['Genotype'])
X_over2, y_over2 = oversampler2.fit_resample(X, y['Treatment_Behavior'])
X_over1_test, y_over1_test = oversampler1_test.fit_resample(X_test, y_test['Genotype'])
X_over2_test, y_over2_test = oversampler2_test.fit_resample(X_test, y_test['Treatment_Behavior'])
```

```python
from sklearn.preprocessing import OneHotEncoder
hot1 = OneHotEncoder()
hot2 = OneHotEncoder()
hot1_test = OneHotEncoder()
hot2_test = OneHotEncoder()
y_over1a = hot1.fit_transform(np.array(y_over1).reshape(-1, 1)).toarray()
y_over2a = hot2.fit_transform(np.array(y_over2).reshape(-1, 1)).toarray()
y_over1a_test = hot1_test.fit_transform(np.array(y_test).reshape(-1, 1)).toarray()
y_over2a_test = hot2_test.fit_transform(np.array(y_test).reshape(-1, 1)).toarray()
```

## Logistic Regression

```python
# for logistic regression
from sklearn.linear_model import LogisticRegression
test_model_lr_binary=LogisticRegression(penalty='elasticnet',solver='saga',l1_ratio=0.5)
test_model_lr_multi=LogisticRegression(penalty='elasticnet',solver='saga',l1_ratio=0.5)
test_model_lr_binary.fit(X_over1, y_over1)
cv_results_lr_binary = cross_validate(test_model_lr_binary, X_over1, y_over1, cv=5)
importances_binary = pd.DataFrame(data={
    'Attribute': X_over1.columns,
    'Importance': test_model_lr_binary.coef_[0]
})
importances_binary = importances_binary.sort_values(by='Importance', ascending=False)
test_model_lr_multi.fit(X_over1, y_over1)
cv_results_lr_multi = cross_validate(test_model_lr_multi, X_over2, y_over2, cv=5)
importances_multi = pd.DataFrame(data={
    'Attribute': X_over1.columns,
    'Importance': test_model_lr_multi.coef_[0]
})
importances_multi = importances_multi.sort_values(by='Importance', ascending=False)

print("Cross Val Score for Binary Classification:", cv_results_lr_binary)
print("Cross Val Score for Multi Classification:", cv_results_lr_multi)
# np.sqrt(np.square(cv_results_lr_binary['test_score']).sum())
# np.sqrt(np.square(cv_results_lr_multi['test_score']).sum())
# Plots
fig = plt.figure(figsize=(20, 3))
plt.bar(x=importances_binary['Attribute'][:35], height=importances_binary['Importance'][:35], color='#087E8B')
plt.title('Feature importances obtained from coefficients for binary classification', size=30)
plt.xticks(rotation='vertical')
plt.show()
fig = plt.figure(figsize=(20, 3))
plt.bar(x=importances_binary['Attribute'][35:], height=importances_binary['Importance'][35:], color='#087E8B')
plt.title('Feature importances obtained from coefficients for binary classification', size=30)
plt.xticks(rotation='vertical')
plt.show()
fig = plt.figure(figsize=(20, 3))
plt.bar(x=importances_multi['Attribute'][:35], height=importances_multi['Importance'][:35], color='#087E8B')
plt.title('Feature importances obtained from coefficients for multi classification', size=30)
plt.xticks(rotation='vertical')
plt.show()
fig = plt.figure(figsize=(20, 3))
plt.bar(x=importances_multi['Attribute'][35:], height=importances_multi['Importance'][35:], color='#087E8B')
plt.title('Feature importances obtained from coefficients for multi classification', size=30)
plt.xticks(rotation='vertical')
plt.show()
```

## Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import cross_val_score
from keras.wrappers.scikit_learn import KerasClassifier
import tensorflow as tf

def getModel():
  model_nn = Sequential()
  model_nn.add(Dense(32, input_dim=74, activation='relu'))
  model_nn.add(Dense(16, activation='relu'))
  model_nn.add(Dense(2, activation='softmax'))
  model_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
  return model_nn
def getModel_multi():
  model_nn = Sequential()
  model_nn.add(Dense(32, input_dim=74, activation='relu'))
  model_nn.add(Dense(16, activation='relu'))
  model_nn.add(Dense(4, activation='softmax'))
  model_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
  return model_nn

ds = X_over1.to_numpy()
ds1 = X_over2.to_numpy()
Kmodel_binary = KerasClassifier(build_fn=getModel, verbose=1)
Kmodel_binary.fit(ds, y_over1a, epochs=20, batch_size=32)
# cv_results_nn = cross_validate(Kmodel, ds, y_over1a, cv=5, scoring=('r2'))
nn_binary_cv_score = cross_val_score(Kmodel_binary, ds, y_over1a, cv=5)

Kmodel_multi = KerasClassifier(build_fn=getModel_multi, verbose=1)
Kmodel_multi.fit(ds1, y_over2a, epochs=20, batch_size=32)
# cv_results_nn = cross_validate(Kmodel_multi, ds1, y_over2a, cv=5, scoring=('r2'))
nn_multi_cv_score = cross_val_score(Kmodel_multi, ds1, y_over2a, cv=5)
print("Cross Val Score for Binary Classification:", nn_binary_cv_score)
print("Cross Val Score for Multi Classification:", nn_multi_cv_score)
# np.sqrt(np.square(nn_binary_cv_score['test_score']).sum())
# np.sqrt(np.square(nn_multi_cv_score['test_score']).sum())
```

# Recursive Feature Elimination


```python
from sklearn.feature_selection import RFECV
model_rfecv = RandomForestClassifier(n_estimators=500)
rfecv1 = RFECV(estimator=model_rfecv,step=1,cv=5,n_jobs=-1)
rfecv1.fit(X_over1,y_over1)
rfecv1.transform(X_over1)
print("For Output 1")
print(rfecv1)
print('Optimal no. of features {}'.format(rfecv1.n_features_))
print(np.where(rfecv1.support_==False)[0])

model_rfecv_2 = RandomForestClassifier(n_estimators=400)
rfecv2 = RFECV(estimator=model_rfecv,step=1,cv=5,n_jobs=-1)
rfecv2.fit(X_over2,y_over2)
rfecv2.transform(X_over2)
print("For Output 2")
print(rfecv2)
print('Optimal no. of features {}'.format(rfecv2.n_features_))
print(np.where(rfecv2.support_==False)[0])
```
