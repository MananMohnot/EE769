```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import sklearn
```

```python
url_cluster="https://www.ee.iitb.ac.in/~asethi/Dump/DataClustering.csv" #referencing url of data
df_cluster = pd.read_csv(url_cluster) #importing train data
display(df_cluster) # display train data
```

```python
pd.plotting.scatter_matrix(df_cluster)
```

```python
df_transformed = np.log(df_cluster) #log transformation
```

```python
pd.plotting.scatter_matrix(df_transformed)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5, random_state=0).fit(df_transformed)
predicted_clusters = kmeans.fit_predict(df_transformed)
pd.plotting.scatter_matrix(df_transformed,c=predicted_clusters,cmap='rainbow')
```

```python
from sklearn.manifold import TSNE
```

```python
# Create a two dimensional t-SNE projection of the embeddings
tsne = TSNE(n_components=2, verbose=1, perplexity=80, n_iter=5000, learning_rate='auto')
tsne_proj = tsne.fit_transform(df_transformed)
tsne_df = pd.DataFrame(tsne_proj, columns=['tsne1', 'tsne2'])
```

```python
pd.plotting.scatter_matrix(tsne_df,c=predicted_clusters,cmap='rainbow')
```
